Compreender como variam os dados de empregos de √°reas de tecnologia para ser poss√≠vel planejar a carreira no futuro do trabalho.

## **‚¨áÔ∏è IMPORTS:**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## **üß© DATASET:**

````python
# Importando e carregando o dataset
url = 'https://raw.githubusercontent.com/yRaffa/GS-2025-2_DataScience/refs/heads/main/data_science_salaries.csv'
df = pd.read_csv(url)

print('Salarios de Data Science:')
df.head()
````

# **üîé AN√ÅLISE DESCRITIVA E EXPLORAT√ìRIA:**

## **üí≤ M√âTRICAS B√ÅSICAS SALARIAIS:**

````python
# C√°lculo de m√©tricas de sal√°rio (m√©dia, mediana e desvio-padr√£o)
media_salario = df['salary_in_usd'].mean()
mediana_salario = df['salary_in_usd'].median()
desvio_padrao_salario = df['salary_in_usd'].std()

df_metricas = pd.DataFrame({
    'M√©trica': ['M√©dia', 'Mediana', 'Desvio-Padr√£o'],
    'Valor (USD)': [f'{media_salario:.2f}', f'{mediana_salario:.2f}', f'{desvio_padrao_salario:.2f}']
})

# DataFrame das m√©tricas de sal√°rio
print('M√©tricas B√°sicas de Sal√°rio:')
display(df_metricas)

# Grafico comparativo das m√©tricas de sal√°rio
metricas_nomes = ['M√©dia', 'Mediana', 'Desvio-padr√£o']
metricas_valores = [media_salario, mediana_salario, desvio_padrao_salario]
plt.figure(figsize=(12,6))
plt.bar(metricas_nomes, metricas_valores)
plt.title('Compara√ß√£o das Estat√≠sticas de Sal√°rio')
plt.ylabel('Valor em USD')
plt.xlabel('M√©tricas')
plt.grid(axis='y', linestyle='--', alpha=0.25)
plt.show()
````

### **üìä M√©tricas B√°sicas Salariais:**

Nesta etapa realizamos uma an√°lise quantitativa da coluna `salary_in_usd` do dataset, calculando as principais m√©tricas de tend√™ncia central e dispers√£o:

- **M√©dia** 
- **Mediana** 
- **Desvio-Padr√£o** 

Essas m√©tricas permitem entender como os sal√°rios est√£o distribu√≠dos e se h√° grande variabilidade dentro do mercado de Data Science.

### **üîç Resultados Interpretados:**

**1. M√©dia Salarial ($145.560,56):**

Representa o sal√°rio m√©dio dos profissionais presentes no dataset. Valores muito altos dentro de determinadas categorias (Senior, Especialistas em IA, ML Engineers) tendem a aumentar esse valor.

**2. Mediana Salarial ($138.666,00):**

Representa o ponto onde 50% dos sal√°rios est√£o acima e 50% abaixo. √â mais confi√°vel do que a m√©dia quando existem muitos outliers e costuma refletir melhor a realidade da maior parte do mercado.

**3. Desvio-Padr√£o da Distribui√ß√£o Salarial ($70.946,84):**

Representa o quanto os sal√°rios variam dentro do conjunto de dados. Um desvio-padr√£o elevado indica que existem grandes diferen√ßas de remunera√ß√£o entre pa√≠ses, n√≠veis de experi√™ncia e cargos.

### **üìå Conclus√£o:**

Os resultados mostram que a distribui√ß√£o de sal√°rios se enquadra em uma **assimetria positiva** `(m√©dia maior que a mediana)`, o que indica a presen√ßa de sal√°rios muito altos que elevam a m√©dia geral. O **desvio-padr√£o elevado** confirma que o mercado √© bastante heterog√™neo.


## **üíØ DISTRIBUI√á√ÉO PERCENTUAL DE VARI√ÅVEIS CATEG√ìRICAS:**

````python
# Distribui√ß√£o percentual das vari√°veis categ√≥ricas (experience_level, employment_type, company_size)
nivel_experiencia = df['experience_level'].value_counts(normalize=True) * 100
df_nivel_experiencia = pd.DataFrame({
    'N√≠vel de Experi√™ncia': list(nivel_experiencia.index),
    'Percentual (%)': nivel_experiencia.round(2).map(lambda x: f'{x:.2f}').tolist()
})

tipo_contrato = df['employment_type'].value_counts(normalize=True) * 100
df_tipo_contrato = pd.DataFrame({
    'Tipo de Contrato': list(tipo_contrato.index),
    'Percentual (%)': tipo_contrato.round(2).map(lambda x: f'{x:.2f}').tolist()
})

porte_empresa = df['company_size'].value_counts(normalize=True) * 100
df_porte_empresa = pd.DataFrame({
    'Porte da Empresa': list(porte_empresa.index),
    'Percentual (%)': porte_empresa.round(2).map(lambda x: f'{x:.2f}').tolist()
})

# DataFrames das distribui√ß√µes percentuais
print('Distribui√ß√£o por N√≠vel de Experi√™ncia:')
display(df_nivel_experiencia)

print('\nDistribui√ß√£o por Tipo de Contrato:')
display(df_tipo_contrato)

print('\nDistribui√ß√£o por Porte da Empresa:')
display(df_porte_empresa)

# Gr√°ficos para as distribui√ß√µes percentuais
fig, axes = plt.subplots(1, 3, figsize=(20,6))

axes[0].pie(nivel_experiencia, labels=nivel_experiencia.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Distribui√ß√£o: N√≠vel de Experi√™ncia')

axes[1].bar(tipo_contrato.index, tipo_contrato.values)
axes[1].set_title('Distribui√ß√£o: Tipo de Contrato')

axes[2].pie(porte_empresa, labels=porte_empresa.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Distribui√ß√£o: Porte da Empresa')

plt.show()
````

### **üìä Distribui√ß√£o Percentual das Vari√°veis Categ√≥ricas**

Para entender a composi√ß√£o do dataset, analisamos a distribui√ß√£o percentual das vari√°veis: `experience_level` `employment_type` `company_size`

### **üîç Resultados Interpretados:**

**1. N√≠vel de Experi√™ncia:**
    
Mostra que no dataset o nivel de experiencia predominante e o N√≠vel-Senior (62.21%).

**2. Tipo de Emprego:**
    
Mostra que o dataset tem majoritariamente contratos Full-Time (99.29%).

**3. Porte da Empresa:**
    
Mostra que a maioria das empresas presentes nesse dataset s√£o de M√©dio Porte (88.80%).

### **üìå Conclus√£o:**

Ao entender essas distribui√ß√µes podemos interpretar m√©tricas salariais mais facilmente, pois temos em mente quais as categorias est√£o mais presentes nas an√°lies, ja que as categoria tem comportamentos diferentes entre si.


## **üèÜ TOP 5 CARGOS COM MAIOR M√âDIA SALARIAL:**

````python
# Top 5 cargos com maior m√©dia salarial
top5 = (
    df.groupby('job_title')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .round(2)
)

df_top5 = pd.DataFrame({
    'Cargo': list(top5.index),
    'M√©dia Salarial (USD)': top5.map(lambda x: f'{x:.2f}').tolist()
})

# DataFrame do Top 5 cargos com maior m√©dia salarial
print('Top 5 Cargos com Maior M√©dia Salarial:')
display(df_top5)

# Gr√°fico do Top 5 cargos com maior m√©dia salarial
plt.figure(figsize=(12,6))
plt.barh(df_top5['Cargo'], df_top5['M√©dia Salarial (USD)'].astype(float))
plt.xlabel("Sal√°rio M√©dio (USD)")
plt.title("Top 5 Cargos com Maior M√©dia Salarial")
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.25)
plt.show()
````

### **üìä Top 5 Cargos com Maior M√©dia Salarial**

Agora analisamos quais cargos apresentam as maiores m√©dias salariais dentro do dataset.

### **üîç Resultados Interpretados:**

Os cargos que lideram o ranking geralmente se destacam por:

- **Fun√ß√µes Estrat√©gicas:** Cargos de lideran√ßa lideram o ranking de m√©dia salarial (Manager, Lead, Architect).
- **Alta Especializa√ß√£o T√©cnica:** √Åreas que dependem de mais especializa√ß√£o t√©cnica tamb√©m se destacam (Machine Learning, IA, Cloud).
- **Demandas Emergentes no Mercado Global:** Complementando os outros resultados, √°reas com demanda emergente possuem uma alta salarial.

Podemos destacar que as √°reas que demandam de especializa√ß√£o t√©cnica tamb√©m s√£o emergentes no mercado.

### **üìå Conclus√£o:**

O Top 5 revela **quais √°reas oferecem os maiores retornos financeiros** e tende a refletir as tend√™ncias atuais do mercado de tecnologia.


## **üîó CORRELA√á√ÉO ENTRE VARI√ÅVEIS CATEG√ìRICAS:**

````python
# Correla√ß√£o entre sal√°rio, taxa de trabalho remoto e porte da empresa
model_map = {'On-site': 1, 'Hybrid': 2, 'Remote': 3}
size_map = {'Small': 1, 'Medium': 2, 'Large': 3}

df['remote_ratio'] = df['work_models'].map(model_map)
df['company_size_num'] = df['company_size'].map(size_map)
correlacao_colunas = ['salary_in_usd', 'remote_ratio', 'company_size_num']
matrix_correlacao = df[correlacao_colunas].corr()

# Renomeando √≠ndices e colunas para melhor apresenta√ß√£o
matrix_correlacao = matrix_correlacao.rename(
    index = {'salary_in_usd': 'Sal√°rio (USD)', 'remote_ratio': 'Trabalho Remoto (%)', 'company_size_num': 'Porte da Empresa'},
    columns = {'salary_in_usd': 'Sal√°rio (USD)', 'remote_ratio': 'Trabalho Remoto (%)', 'company_size_num': 'Porte da Empresa'}
)

correlaco_plot = matrix_correlacao.copy()
correlaco_plot.index = ["Sal√°rio (USD)", "Trabalho Remoto (%)", "Porte da Empresa"]
correlaco_plot.columns = ["Sal√°rio (USD)", "Trabalho Remoto (%)", "Porte da Empresa"]

# DataFrame da matriz de correla√ß√£o
print('Matriz de Correla√ß√£o: Sal√°rio √ó Taxa de Trabalho Remoto √ó Porte da Empresa:')
display(matrix_correlacao)

# Gr√°fico da matriz de correla√ß√£o
plt.figure(figsize=(12,6))
sns.heatmap(correlaco_plot, annot=True, cmap="Blues", linewidths=.5, fmt=".2f")
plt.title('Matriz de Correla√ß√£o: Sal√°rio √ó Taxa de Trabalho Remoto √ó Porte da Empresa')
plt.show()
````

### **üìä Correla√ß√£o entre Vari√°veis Categ√≥ricas**

Para investigar poss√≠veis rela√ß√µes entre **Sal√°rio √ó Taxa de Trabalho Remoto √ó Porte da Empresa**, analisamos a correla√ß√£o entre as vari√°veis: `salary_in_usd` `remote_ratio` `company_size`

O Heatmap mostra a for√ßa e dire√ß√£o das correla√ß√µes.

- Valores **Pr√≥ximos de +1** ‚Üí Correla√ß√£o positiva forte.
- Valores **Pr√≥ximos de -1** ‚Üí Correla√ß√£o negativa forte.
- Valores **Pr√≥ximos de 0** ‚Üí Pouca ou nenhuma correla√ß√£o linear.

### **üîç Resultados Interpretados:**

**1. Sal√°rio √ó Porte da Empresa ‚Üí -0.09**

- Correla√ß√£o quase **nula** e **levemente negativa**.

Empresas grandes n√£o pagam maiores sal√°rios que empresas pequenas, nem o contr√°rio.

**2. Sal√°rio √ó Taxa de Trabalho Remoto ‚Üí -0.03**
   
- Correla√ß√£o quase **nula** e **levemente negativa**.

Trabalhar de forma remota n√£o aumenta nem diminui o sal√°rio de forma significativa.

**3. Taxa de Trabalho Remoto √ó Porte da Empresa ‚Üí 0.04**
   
- Correla√ß√£o quase **nula** e **levemente positiva**.

Empresas maiores podem ter uma tend√™ncia a mais trabalhos remotos, mas n√£o √© regra.

### **üìå Conclus√£o:**

N√£o existe correla√ß√£o linear forte entre **Sal√°rio √ó Taxa de Trabalho Remoto √ó Porte da Empresa**. O dataset mostra um cen√°rio, onde o cargo, a senioridade, a regi√£o e a √°rea espec√≠fica provavelmente s√£o mais significativos para o sal√°rio.


## **üìà BOXPLOT DE SAL√ÅRIO √ó N√çVEL DE EXPERI√äNCIA E LOCALIZA√á√ÉO DA EMPRESA:**

````python
# Gr√°fico Boxplot de Sal√°rio por N√≠vel de Experi√™ncia
plt.figure(figsize=(16,6))
sns.boxplot(data=df, x="experience_level", y="salary_in_usd")
plt.title("Distribui√ß√£o de Sal√°rio por N√≠vel de Experi√™ncia")
plt.xlabel("N√≠vel de Experi√™ncia")
plt.ylabel("Sal√°rio (USD)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.25)
plt.show()

# Gr√°fico Boxplot de Sal√°rio por Localiza√ß√£o da Empresa (Top 10 em n¬∫ de vagas)
top_countries = df["company_location"].value_counts().head(10).index
df_top_countries = df[df["company_location"].isin(top_countries)]
plt.figure(figsize=(16,6))
sns.boxplot(data=df_top_countries, x="company_location", y="salary_in_usd")
plt.title("Distribui√ß√£o de Sal√°rio por Localiza√ß√£o da Empresa (Top 10 em n¬∫ de Vagas)")
plt.xlabel("Localiza√ß√£o da Empresa")
plt.ylabel("Sal√°rio (USD)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.25)
plt.show()
````
